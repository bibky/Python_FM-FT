```python
#Задание 2
from collections import defaultdict, ChainMap
from typing import Union
Books = dict[str, defaultdict]
archive: Books = {letter: defaultdict(lambda: defaultdict(list)) for letter in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЭЮЯ'}
def  addition_marks(surname: str, category: str, object: Union [str, list]) -> dict:
    first_letter = surname[0].upper()
    book = archive[first_letter]
    book[surname][category].extend(object)

def search_category(category: str) -> ChainMap:
    result = ChainMap()
    for letter in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЭЮЯ':
        book = archive[letter]
        for surname, categories in book.items():
            if category in categories:
                result = result.new_child({surname: categories[category]})
    return result
furniture = search_category("мебель")
addition_marks("Иванов", "мебель", ["стол", "стул"])
addition_marks("Петров", "посуда", "тарелка")
addition_marks("Иванов", "посуда", ["чашка", "ложка"])
addition_marks("Сидоров", "мебель", "кровать")
for surname, items in furniture.items():
        print(f"{surname}: {items}")

print("\n Книга на букву 'И':")
for surname, categories in archive['И'].items():
    print(f"{surname}: {dict(categories)}")


```

```python
import sys
import os
import time
from collections import defaultdict

def get_directory_size(directory):
    """Вычисляет размер директории в байтах"""
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            try:
                total_size += os.path.getsize(filepath)
            except (OSError, FileNotFoundError):
                continue
    return total_size

def format_size(size_bytes):
    """Форматирует размер в удобочитаемый вид"""
    for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
        if size_bytes < 1024.0:
            return f"{size_bytes:.2f} {unit}"
        size_bytes /= 1024.0
    return f"{size_bytes:.2f} PB"

def analyze_directories():
    """Анализирует директории и выводит их с размерами, отсортированные по времени"""
    print("=" * 80)
    print("ДИРЕКТОРИИ С РАЗМЕРАМИ ФАЙЛОВ (отсортированы по времени последнего обращения)")
    print("=" * 80)
    
    # Получаем текущую директорию и ее родительскую
    current_dir = os.getcwd()
    home_dir = os.path.expanduser("~")
    
    # Собираем информацию о директориях
    directories_info = []
    
    # Анализируем несколько важных директорий
    target_dirs = [
        current_dir,
        home_dir,
        os.path.join(home_dir, "Desktop"),
        os.path.join(home_dir, "Documents"),
        os.path.join(home_dir, "Downloads"),
        os.path.join(home_dir, "Pictures"),
    ]
    
    for directory in target_dirs:
        if os.path.exists(directory) and os.path.isdir(directory):
            try:
                size = get_directory_size(directory)
                access_time = os.path.getatime(directory)
                directories_info.append({
                    'path': directory,
                    'size': size,
                    'access_time': access_time
                })
            except (OSError, PermissionError):
                continue
    
    # Сортируем по времени последнего обращения (новые сначала)
    directories_info.sort(key=lambda x: x['access_time'], reverse=True)
    
    # Выводим результаты
    for info in directories_info:
        access_time_str = time.strftime('%Y-%m-%d %H:%M:%S', 
                                      time.localtime(info['access_time']))
        print(f"Директория: {info['path']}")
        print(f"Размер: {format_size(info['size'])}")
        print(f"Последнее обращение: {access_time_str}")
        print("-" * 80)

def find_file_extensions():
    """Находит все уникальные расширения файлов на диске"""
    print("\n" + "=" * 80)
    print("СПИСОК РАСШИРЕНИЙ ФАЙЛОВ НА ВАШЕМ ПК")
    print("=" * 80)
    
    extensions = set()
    home_dir = os.path.expanduser("~")
    
    # Ограничим глубину поиска для производительности
    max_files = 10000
    file_count = 0
    
    for dirpath, dirnames, filenames in os.walk(home_dir):
        for filename in filenames:
            if file_count >= max_files:
                break
                
            # Получаем расширение файла
            _, ext = os.path.splitext(filename)
            if ext:  # Игнорируем файлы без расширения
                extensions.add(ext.lower())
            
            file_count += 1
        
        if file_count >= max_files:
            break
    
    # Выводим расширения в алфавитном порядке
    sorted_extensions = sorted(extensions)
    print(f"Найдено расширений: {len(sorted_extensions)}")
    print("Расширения:", ", ".join(sorted_extensions))

def calculate_interpreter_memory():
    """Вычисляет объем памяти, используемый интерпретатором Python"""
    print("\n" + "=" * 80)
    print("ИНФОРМАЦИЯ О ПАМЯТИ ИНТЕРПРЕТАТОРА PYTHON")
    print("=" * 80)
    
    try:
        # Используем модуль psutil для более точного измерения памяти
        import psutil
        process = psutil.Process()
        memory_info = process.memory_info()
        
        print(f"Используемая память: {format_size(memory_info.rss)}")
        print(f"Виртуальная память: {format_size(memory_info.vms)}")
        
    except ImportError:
        print("Для точного измерения памяти установите модуль psutil:")
        print("pip install psutil")
        
        import resource
        memory_usage = resource.getrusage(resource.RUSAGE_SELF).ru_maxrss
        if os.name == 'posix':  # Linux/Mac
            memory_usage *= 1024  # Конвертируем из KB в байты
        
        print(f"Примерный объем памяти: {format_size(memory_usage)}")

def main():
    """Основная функция"""
    print("АНАЛИЗ СИСТЕМЫ")
    print("=" * 80)
    
    # Информация о системе
    print(f"Платформа: {sys.platform}")
    print(f"Версия Python: {sys.version}")
    print(f"Текущая рабочая директория: {os.getcwd()}")
    print()
    
    # Выполняем задачи
    analyze_directories()
    find_file_extensions()
    calculate_interpreter_memory()
    
    # Дополнительная информация
    print("\n" + "=" * 80)
    print("ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ")
    print("=" * 80)
    print(f"Размер sys модуля: {format_size(sys.getsizeof(sys))}")
    print(f"Размер os модуля: {format_size(sys.getsizeof(os))}")
    
    # Информация о файлах Python
    python_exe = sys.executable
    if os.path.exists(python_exe):
        size = os.path.getsize(python_exe)
        print(f"Размер интерпретатора Python: {format_size(size)}")

if __name__ == "__main__":
    main()
```
